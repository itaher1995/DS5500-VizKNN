```python
# imports
import pandas as pd
import numpy as np
from scipy.stats import boxcox
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import f1_score
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.feature_selection import mutual_info_classif
from sklearn.decomposition import PCA
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot  as plt
import seaborn as sns
%matplotlib inline
```

## Data Preprocessing

```python
data = pd.read_csv('adult.csv',header=None)
data.columns = ['age','workclass','fnlwgt','education','education_num','martial_status','occupation','relationship','race','sex','capital_gain','capital_loss','hours_per_week','native_country','salary']
data.drop('education_num',axis=1,inplace=True)
data['salary'] = [1 if x==' >50K' else 0 for x in data['salary']]
data = pd.get_dummies(data) # dummy variables for categorical data
X_train, X_test, y_train, y_test = train_test_split(data.drop('salary',axis=1),data['salary'],test_size=0.2)
X_train
```

```python
# initialize min-max scalers for transformation

ageMinMax = MinMaxScaler()
fnlwgtMinMax = MinMaxScaler()
capGainMinMax = MinMaxScaler()
capLossMinMax = MinMaxScaler()
hpwMinMax = MinMaxScaler()

# fit min max scalers to respective columns

ageMinMax.fit(boxcox(X_train['age'].values.reshape(-1, 1))[0]) # use boxcox to reduce skew in data
fnlwgtMinMax.fit(boxcox(X_train['fnlwgt'].values.reshape(-1, 1))[0]) # use boxcox to reduce skew in data
capGainMinMax.fit(X_train['capital_gain'].values.reshape(-1, 1))
capLossMinMax.fit(X_train['capital_loss'].values.reshape(-1, 1))
hpwMinMax.fit(X_train['hours_per_week'].values.reshape(-1, 1))

# transform the training and test data given this information
X_train['age'] = ageMinMax.transform(boxcox(X_train['age'].values.reshape(-1, 1))[0])
X_test['age'] = ageMinMax.transform(boxcox(X_test['age'].values.reshape(-1, 1))[0])

X_train['fnlwgt'] = fnlwgtMinMax.transform(boxcox(X_train['fnlwgt'].values.reshape(-1, 1))[0])
X_test['fnlwgt'] = fnlwgtMinMax.transform(boxcox(X_test['fnlwgt'].values.reshape(-1, 1))[0])

X_train['capital_gain'] = capGainMinMax.transform(X_train['capital_gain'].values.reshape(-1, 1))
X_test['capital_gain'] = capGainMinMax.transform(X_test['capital_gain'].values.reshape(-1, 1))

X_train['capital_loss'] = capLossMinMax.transform(X_train['capital_loss'].values.reshape(-1, 1))
X_test['capital_loss'] = capLossMinMax.transform(X_test['capital_loss'].values.reshape(-1, 1))

X_train['hours_per_week'] = hpwMinMax.transform(X_train['hours_per_week'].values.reshape(-1, 1))
X_test['hours_per_week'] = hpwMinMax.transform(X_test['hours_per_week'].values.reshape(-1, 1))
```

For the data preprocessing I followed four very simple steps.

1. I dropped education_num as it is just a numerical encoding of the education variable

2. For all categorical values, I created dummy variables so that they could be used in our classifiers. 

3. For the age and fnlwght variables, there was noticeable skew in the data. In order to reduce that skew, I tried the log transformation and the boxcox transformation and found qualitatively that the boxcox attempted to transform the distribution towards being more normal. This is supported by the empirical use case of boxcox. While it does not gurantee a normal distribution, the boxcox transformation can be empirically used to create normal distributions from non-normal data. In the scipy implementation of boxcox, there is a parameter, lambda, that if it is not set, the function will attempt to maximize the log-likelihood w.r.t. lambda. While the log function can also reduce skew, the box-cox is a generalized data transformation function. It can replicate the effects of the log transform if lambda is set to 0.5.

4. Finally, in order to reduce the bias variables of different magnitudes might cause, I used a min-max scaler fit on the training data. Then, I transformed the training and test data using the given scalers.

I wanted to visualize the various distributions found in the model by class. Thus I created violin plots for each continuous variable in the dataset and labeled the classes. The idea is to look for differences in the distributions to see if they alone could discriminate between classes. It seems as though the distribution for the age of individuals who make more than 50K has its distribution shifted to the right after applying the log function compared to the distribution of ages for individuals who make less than 50K. Intuitively this makes sense, as the older you get one can expect to make more. Therefore, this predictor might be useful in discriminating between classes. The violin plot for hours per week shows a skewed distribution, where people who make more than 50k tend to work more hours, but a large percentage still work a 40hr week. This is also intuitive. This also might be helpful for building a model as higher values might tend to relate to the >50K class.

The type of chart used is a small multiples plot of violin plots. This was used over boxplots because the visualizations shows the data distribution, which would not be encoded using just boxplots [Visual Encodings, 37]. While one can make an argument that one of the channels we are encoding is area, the primary channel here is position. Position is the easiest magnitude channel for users to understand, so that is why I used it [Marks and Channels, 40]. The mark is area. When looking at this plot, I looked primarily for where most of the area of each distribution is on the y-axis. Each small multiple is on a rectilinear axis [Visual Encodings, 27].

### Visualization 2

```python
def plotProp(X,response,num_rows,num_cols,drop_cols,dummyPref,size,selectedCols=None):
    plotData = X.copy(deep=True)
    plotData['response'] = response
    plotData.drop(drop_cols,axis=1,inplace=True)
    plotData = plotData.groupby('response').sum()
    plotData = plotData.apply(lambda x: x/y_train.value_counts())
    plotData.index = ['<=50K' if x==0 else '>50K' for x in plotData.index]

    f, axes = plt.subplots(num_rows, num_cols, figsize=size, sharex=True)
    f.suptitle('Bar Plots of the Proportion of Response')
    sns.despine(left=True)
    x=0
    y=0
    if dummyPref != None:
        cols = [x for x in plotData.columns if x.startswith(dummyPref)]
    else:
        cols = selectedCols

    for col in cols:
        if y==num_cols:
            x+=1
            y=0
        if num_rows>1:
            sns.barplot(x=plotData.index,y=plotData[col],ax=axes[x,y])
        else:
            sns.barplot(x=plotData.index,y=plotData[col],ax=axes[y])
        y+=1

plotProp(X_train,y_train,5,3,['age','fnlwgt','capital_gain','capital_loss','hours_per_week'],'occupation',(20,20))
  
```

This is a small multiples chart of the workclass variable by each each occupation. Small multiples of proportions are used here instead of a stacked bar plot so we don't lose information due to the amount of categories. Intuituvely we see, executives in mangerial positions and special professions constitute larger proportions of the >50K compared to the <=50K. We also see that priv-house-serv and cleaners consitute larger proportions of the <=50K class, even though the difference magnitude is quite small. This information can help build a model because it sheds light on potential further feature engineering. The data suggests that there are some generally higher salary positions.

I encode the length channel here as it is very easy to interpret because for its expressiveness and effectiveness [Marks, 40]. The plot has rectilinear axis [Visual Encodings, 27]. Color is also encoded as an identity variable, though it is encoded redundantly as the categories are on the x-axis. The length channel is appropriate because it allows one to visualize the differences in the magnitude of the proportion each occupation takes up of our response variable classes. 


```python
def occupationFeatureEngineering(data):
    '''
    Add two new features to the data "Professional_Executive" and "NonProfessional", based on prior data visualization.
    '''
    dataCopy = data.copy(deep=True)
    dataCopy['Professional_Executive'] = dataCopy['occupation_ Exec-managerial'] + dataCopy['occupation_ Prof-specialty']
    dataCopy['NonProfessional'] = dataCopy['occupation_ Priv-house-serv'] + dataCopy['occupation_ Other-service'] + dataCopy['occupation_ Handlers-cleaners'] + dataCopy['occupation_ Farming-fishing'] + dataCopy['occupation_ ?'] + dataCopy['occupation_ Machine-op-inspct']
    dataCopy.drop(['occupation_ Exec-managerial','occupation_ Prof-specialty','occupation_ Priv-house-serv','occupation_ Other-service','occupation_ Handlers-cleaners','occupation_ Farming-fishing','occupation_ ?','occupation_ Machine-op-inspct'],axis=1,inplace=True)
    return dataCopy

X_trainNew = occupationFeatureEngineering(X_train)
X_testNew = occupationFeatureEngineering(X_test)


plotProp(X_trainNew,y_train,1,2,['age','fnlwgt','capital_gain','capital_loss','hours_per_week'],None,(10,5),['Professional_Executive','NonProfessional'])
    
```

## Modeling

```python
def elbowPointPCA(features,response):
    '''
    Develop the elbow point graph for pca to qualitatively choice the number of components.
    '''
    pca = PCA()
    pca.fit(features,response)
    plt.plot(np.cumsum(pca.explained_variance_ratio_))

elbowPointPCA(X_trainNew,y_train) #n_components is around 45
```

```python
pca = PCA(n_components=45) # Had to use PCA because dimensions were far too high, was increasing training time.
pca.fit(X_trainNew,y_train)
X_trainPCA = pca.transform(X_trainNew)
X_testPCA = pca.transform(X_testNew)
```

```python
gs = GridSearchCV(estimator=KNeighborsClassifier(),param_grid={'n_neighbors':list(range(1,11)),'p':[1,2]},scoring='f1',cv=10)
gs.fit(X_trainPCA,y_train)
```

```python
gsRes = pd.DataFrame(gs.cv_results_)
```

### Visualization 3

```python
gsResPlot = gsRes[['param_n_neighbors']+[x for x in gsRes.columns if 'test' in x and 'split' in x]].loc[gsRes['param_p']==2]
sns.violinplot(data=gsResPlot.melt(id_vars=['param_n_neighbors']),x='param_n_neighbors',y='value')
```

In order to showcase the performance of the model with the l2 distance and k from 1 to 10, I used a violin plot for each k over all 10 cv. I wanted to encode more data regarding the distribution of F1 scores across all splits for each k. Therefore, I encode the position channel in order to compare distributions and the color to identify the distributions for each neighbor. The mark I use is area and point for the median of the violin plot. Using this information, it is easy to tease out that 1 and 2 neighbors was too low. We see that the highest medians are found at n_neighbors equal 9, which might point to k=9 being the best parameter when the distance is fixed.
